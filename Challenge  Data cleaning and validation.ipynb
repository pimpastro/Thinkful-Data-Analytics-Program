## Question 3:
bonus round, identify the open access prices paid by subject area

```python

l= []

for i , row in df.iterrows():

    if 'biology' in row['Publisher'] or 'biology' in row['Journal title'] or 'biology' in row['Article title']:
        l.append('biology')
        
    elif 'chemistry' in row['Publisher'] or 'chemistry' in row['Journal title'] or 'chemistry' in row['Article title']:
        l.append('chemistry')
        
    elif 'immunology' in row['Publisher'] or 'immunology' in row['Journal title'] or 'immunology' in row['Article title']:
        l.append('immunology')
        
    elif 'aids' in row['Publisher'] or 'aids' in row['Journal title'] or 'aids' in row['Article title']:
        l.append('aids')
        
    else:
        l.append('other')
        

        
df['sub']= l
```
