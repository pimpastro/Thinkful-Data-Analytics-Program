Reproduce the pop1 and pop2 populations and samples using numpy's binomial function. 

Specifically, create two binomially distributed populations with n equal to 10 and size equal to 10000. 

The p-value of pop1 should be 0.2 and the p-value of pop2 should be 0.5. 

Using a sample size of 100, calculate the means and standard deviations of your samples.


```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
%matplotlib inline
```

```python
# Creating two binomially distributed populations with n equal to 10 and size equal to 10000.
# The p-value of pop1 = 0.2 and the p-value of pop2 = 0.5.
pop1 = np.random.binomial(10, 0.2, 10000)
pop2 = np.random.binomial(10,0.5, 10000) 
```

```python
# Letâ€™s make histograms for the two groups. 

plt.hist(pop1, alpha=0.5, label='Population 1') 
plt.hist(pop2, alpha=0.5, label='Population 2') 
plt.legend(loc='upper right') 
plt.show()
```

```python
# Creating a samples
sample1 = np.random.choice(pop1, 100, replace=True)
sample2 = np.random.choice(pop2, 100, replace=True)

mean_sample1 = np.mean(sample1)
st_dev_sample1 = np.std(sample1)


plt.hist(sample1, alpha=0.5, label='sample 1') 
plt.hist(sample2, alpha=0.5, label='sample 2') 

mean_sample2 = np.mean(sample2)
st_dev_sample2 = np.std(sample2)

plt.axvline(x=mean_sample1, linestyle='solid', color='black')
plt.axvline(x=mean_sample1+st_dev_sample1, linestyle='dashed', color='black')
plt.axvline(x=mean_sample1-st_dev_sample1, linestyle='dashed', color='black')

plt.axvline(x=mean_sample2, linestyle='solid', color='blue')
plt.axvline(x=mean_sample2+st_dev_sample2, linestyle='dashed', color='blue')
plt.axvline(x=mean_sample2-st_dev_sample2, linestyle='dashed', color='blue')

plt.legend(loc='upper right') 
plt.show()

# Calculate the means and standard deviations of our samples.

print(sample1.mean())
print(sample2.mean())
print(sample1.std())
print(sample2.std())

# Compute the difference between the two sample means.
diff=sample2.mean( ) -sample1.mean()
print(diff)

print(ttest_ind(sample2, sample1, equal_var=False))
```

Change the distribution of your populations from binomial to a distribution of your choice. Do the sample mean values still accurately represent the population values?

```python
# The Central Limit Theorem says the statistics should still work as long as the sample size is large enough, 
# no matter what distribution is chosen
```

```python
population_1 = np.random.exponential(scale=1.0, size=100)
population_2 = np.random.exponential(scale=1.0, size=100)
plt.hist(population_1, alpha=0.5, label='Exponential Population 1')
plt.hist(population_2, alpha=0.2, label='Exponential Population 2')  


plt.legend(loc='upper right') 
plt.show()
```

```python
sample_exp_1 = np.random.choice(population_1, 100, replace=True)
sample_exp_2 = np.random.choice(population_2, 100, replace=True)

mean_sample1_exp_1 = np.mean(sample_exp_1)
st_dev_sample1_exp_1 = np.std(sample_exp_1)

plt.axvline(x=mean_sample1_exp_1, linestyle='solid', color='black')
plt.axvline(x=mean_sample1_exp_1+st_dev_sample1_exp_1, linestyle='dashed', color='black')
plt.axvline(x=mean_sample1_exp_1-st_dev_sample1_exp_1, linestyle='dashed', color='black')

mean_sample_exp_2 = np.mean(sample_exp_2)
st_dev2_sample_exp_2 = np.std(sample_exp_2)

plt.axvline(x=mean_sample_exp_2, linestyle='solid', color='blue')
plt.axvline(x=mean_sample_exp_2+st_dev2_2, linestyle='dashed', color='blue')
plt.axvline(x=mean_sample_exp_2-st_dev2_2, linestyle='dashed', color='blue')

plt.hist(sample_exp_1, alpha=0.5, label='sample_exp_1') 
plt.hist(sample_exp_2, alpha=0.5, label='sample_exp_2') 
plt.legend(loc='upper right') 
plt.show()

print('mean_sample_exp_1.mean:', sample_exp_1.mean())
print('mean_sample_exp_2.mean:', sample_exp_2.mean())
print('st_dev2_sample_exp_1:', sample_exp_1.std())
print('st_dev2_sample_exp_2:', sample_exp_2.std())
# Compute the difference between the two sample means.
diff_2 =sample_exp_2.mean( ) - sample_exp_1.mean()
print(diff_2)

print(ttest_ind(sample_exp_2, sample_exp_1, equal_var=False))
```
